Sabemos la ecuacin de la magnitud aparente:
$$m_{x}=-2.5 \log _{10}\left(\frac{F_{x}}{  F_{x}^{0}}\right)$$


Despejando para $F_x$ (flujo): 

$$m_{x}=-2.5 [\log _{10}F_{x}-\log_{10}F_x^0]$$
$$-\frac{m_x}{2.5}=\log _{10}F_{x}-\log_{10}F_x^0 $$
$$-\frac{m_x}{2.5}+\log_{10}F_x^0=\log _{10}F_{x} $$
$$10^{-\frac{m_x}{2.5}+\log_{10}F_x^0}=10^{\log _{10}F_{x}}$$
$$F_x=10^{-\frac{m_x}{2.5}+\log_{10}F_x^0}$$
$$F_x=10^{-\frac{m_x}{2.5}}\cdot F_x^0$$

Obteniendo el flujo base:
\begin{align} F_x^0&= 4.283\cdot 10^{-14} Wcm^{-2}\mu m^{-1}\\
F_x^0&= 4.283\cdot 10^{-14}\cdot 10^{-2}\cdot 10^{-6} Wm^{-2}m^{-1}\\
F_x^0&= 4.283\cdot 10^{-22} Wm^{-3}\\
F_x^0&= 4.283\cdot 10^{-22} Wm^{-3}
\end{align}

```
Flujo_base = 4.283E-22
Flujo= Flujo_base*10**(-magnitud_aparente/2.5)
print(f"Flujo = {Flujo}")
```

Luminosidad
$$L=FA$$
$$L=F4\pi r^2$$

```
Luminosidad_estrella = (Flujo*4*np.pi)*(distancia_m)**2
print(f"Luminosidad = {Luminosidad_estrella}")
```

Temperatura 
$$L= \sigma AT^4$$
$$T=\left( \frac{L}{\sigma A}\right)^{1/4}$$
$$T=\left( \frac{L}{\sigma 4\pi r^2}\right)^{1/4}$$
Considerando $\sigma=5.670\cdot 10^{-8}Wm^{-2}K^{-4}$

```
sigma= 5.670E-8
lo_de_abajo = sigma*4*np.pi*(distancia_m)**2
Temperatura= (Luminosidad_estrella/lo_de_abajo)**(1/4)
print(f"Temperatura = {Temperatura} Kelvin")
```
