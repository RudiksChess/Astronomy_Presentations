Sabemos la ecuacin de la magnitud aparente:
$$m_{x}=-2.5 \log _{10}\left(\frac{F_{x}}{  F_{x}^{0}}\right)$$


Despejando para $F_x$ (flujo): 

$$m_{x}=-2.5 [\log _{10}F_{x}-\log_{10}F_x^0]$$
$$-\frac{m_x}{2.5}=\log _{10}F_{x}-\log_{10}F_x^0 $$
$$-\frac{m_x}{2.5}+\log_{10}F_x^0=\log _{10}F_{x} $$
$$10^{-\frac{m_x}{2.5}+\log_{10}F_x^0}=10^{\log _{10}F_{x}}$$
$$F_x=10^{-\frac{m_x}{2.5}+\log_{10}F_x^0}$$
$$F_x=10^{-\frac{m_x}{2.5}}\cdot F_x^0$$



Screen Shot 2020-11-25 at 21.51.01.png

Obteniendo el flujo base:
\begin{align} F_x^0&= 4.283\cdot 10^{-14} Wcm^{-2}\mu m^{-1}*2.2\mu m\\
F_x^0&= 9.4226\cdot 10^{-14}\cdot 10^{4} Wm^{-2}\\
F_x^0&= 9.4226\cdot 10^{-10} Wm^{-2}\\
F_x^0&= 9.4226\cdot 10^{-10} Wm^{-2}
\end{align}

```
Flujo_base = 9.4226*10**(-10)
Flujo= Flujo_base*10**(-magnitud_aparente/2.5)
print(f"Flujo = {Flujo}"+" W/m")
```

Luminosidad
$$L=FA$$
$$L=F4\pi r^2$$

Comparado al sol: 
$$\frac{L}{L_{\odot}}$$

Donde $L_{\odot}=3.828\cdot 10^{26}W$

```
Luminosidad_sol =3.828E26
Luminosidad_estrella = (Flujo*4*np.pi)*(distancia_m)**2
Luminosidad_comparado = Luminosidad_estrella/Luminosidad_sol
print(f"Luminosidad = {Luminosidad_comparado} W")
```

Tamao 
 $$F=\frac{L}{A}=\frac{L}{4\pi R^2}=\sigma T^4$$
  $$R=\sqrt{\frac{L}{4\pi \sigma T^4}}$$
Radio del sol = 696340$\cdot 10^3$ m
Radio comparado = $\frac{R}{R_{\odot}}$


```
radio_del_sol = 696340e3
R= np.sqrt(Luminosidad_estrella/(4*np.pi*5.6704e-8 *(temperatura)**4 ))
radio_comparado = R/radio_del_sol
print(f"Radio = {radio_comparado}")
```
